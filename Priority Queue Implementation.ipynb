# Priority Queue Introduction

Basically, a Priority Queue is an abstract data structure where elements aren't processed on a first-come, first-served basis. Instead, each element has a specific priority attached to it, and elements with higher priority are processed before the ones with lower priority. Now, if we try to implement this using normal arrays or linked lists, the time complexity becomes very bad. Therefore, the most efficient way to implement a priority queue is by using a Binary Heap, which easily supports insertion and deletion operations in logarithmic time (O log n)
In this notebook, I have implemented a Max-Heap using a standard Python list to manage various tasks based on their priority values.

Each task is represented by a dedicated Task class containing the following attributes:

    •	Task ID
    •	Priority
    •	Arrival Time
    •	Deadline

The implementation covers all the core priority queue operations required as per our syllabus: insert(), extract_maximum(), increase_key(), and is_empty(). 


```python
class Task:

    def __init__(self, task_id, priority, arrival_time, deadline):

        self.task_id = task_id

        self.priority = priority

        self.arrival_time = arrival_time

        self.deadline = deadline

    def __repr__(self):

        return (f"Task(ID={self.task_id}, "

                f"Priority={self.priority}, "

                f"Arrival={self.arrival_time}, "

                f"Deadline={self.deadline})")
```

```python
class PriorityQueue:

    def __init__(self):
        self.heap = []

    # Parent index
    def parent(self, i):
        return (i - 1) // 2

    # Left child
    def left(self, i):
        return 2 * i + 1

    # Right child
    def right(self, i):
        return 2 * i + 2

    # Insert a new task
    def insert(self, task):

        self.heap.append(task)

        i = len(self.heap) - 1

        while i > 0 and self.heap[self.parent(i)].priority < self.heap[i].priority:

            self.heap[i], self.heap[self.parent(i)] = \
                self.heap[self.parent(i)], self.heap[i]

            i = self.parent(i)

    # Heapify downward
    def heapify(self, i):

        largest = i

        left = self.left(i)
        right = self.right(i)

        if left < len(self.heap) and \
           self.heap[left].priority > self.heap[largest].priority:
            largest = left

        if right < len(self.heap) and \
           self.heap[right].priority > self.heap[largest].priority:
            largest = right

        if largest != i:

            self.heap[i], self.heap[largest] = \
                self.heap[largest], self.heap[i]

            self.heapify(largest)

    # Remove highest-priority task
    def extract_max(self):

        if self.is_empty():
            return None

        maximum = self.heap[0]

        self.heap[0] = self.heap[-1]

        self.heap.pop()

        if not self.is_empty():
            self.heapify(0)

        return maximum

    # Increase priority
    def increase_key(self, task_id, new_priority):

        for i in range(len(self.heap)):

            if self.heap[i].task_id == task_id:

                if new_priority < self.heap[i].priority:
                    return

                self.heap[i].priority = new_priority

                while i > 0 and \
                      self.heap[self.parent(i)].priority < self.heap[i].priority:

                    self.heap[i], self.heap[self.parent(i)] = \
                        self.heap[self.parent(i)], self.heap[i]

                    i = self.parent(i)

                return

    # Check if queue is empty
    def is_empty(self):
        return len(self.heap) == 0
```

```python
# Driver Program

pq = PriorityQueue()

# Create sample tasks
tasks = [
    Task(101, 5, 0, 12),
    Task(102, 8, 2, 15),
    Task(103, 3, 4, 20),
    Task(104, 10, 1, 8),
    Task(105, 6, 3, 18)
]

# Insert tasks
print("Inserting Tasks:")
for task in tasks:
    print(task)
    pq.insert(task)

print("\nCurrent Heap:")
for task in pq.heap:
    print(task)

# Extract highest-priority task
print("\nExtract Maximum:")
print(pq.extract_max())

print("\nHeap after extraction:")
for task in pq.heap:
    print(task)

# Increase priority of Task 103
print("\nIncreasing priority of Task 103 to 12...")
pq.increase_key(103, 12)

print("\nHeap after increasing priority:")
for task in pq.heap:
    print(task)

# Extract all remaining tasks
print("\nExtracting Remaining Tasks:")
while not pq.is_empty():
    print(pq.extract_max())
```

Summary of Results

The priority queue implementation using a max-heap functioned correctly for all required operations. Five sample tasks with different priority values were successfully inserted into the priority queue while maintaining the max-heap property. After all insertions, the task with the highest priority (Task 104 with priority 10) was correctly placed at the root of the heap.

The extract_max() operation successfully removed and returned the highest-priority task, after which the heap was reorganized to maintain the max-heap property. The next highest-priority task (Task 102 with priority 8) became the new root, demonstrating that the heap structure was correctly restored.

The increase_key() operation was then performed by increasing the priority of Task 103 from 3 to 12. After updating the priority, the task moved upward in the heap and became the new root, confirming that the heap property was properly maintained after a priority change.

Finally, repeated calls to extract_max() removed the remaining tasks in descending order of priority: Task 103 (priority 12), Task 102 (priority 8), Task 105 (priority 6), and Task 101 (priority 5). This verifies that the priority queue always returns the task with the highest priority first, as expected for a max-heap implementation.

Overall, the experimental results confirm that the implemented priority queue correctly supports insertion, maximum extraction, priority updates, and empty-queue checking while maintaining the heap property throughout all operations. The observed behavior is consistent with the theoretical time complexities of O(log n) for insertion and extraction, O(n) for the implemented increase_key() operation due to linear search, and O(1) for checking whether the queue is empty.
