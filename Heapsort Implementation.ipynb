# Heapsort Introduction

Heapsort is a comparison-based sorting algorithm that uses a binary heap data structure. The algorithm works by first converting the input array into a max-heap, where the largest element is always stored at the root. Then, it repeatedly extracts the maximum element and places it at the end of the array while restoring the heap property.

The implementation below follows two main steps:

Step 1: First, build a max-heap from the given array.

Step 2: Then, repeatedly remove the max element and dump it into its correct sorted position at the back. Done!

```python
def heapify(arr, n, i):
    
    #Maintains the max-heap property for a subtree rooted at index i.
    
    
    largest = i
    left = 2 * i + 1
    right = 2 * i + 2

    # Checking if left child exists and is greater than root
    
    if left < n and arr[left] > arr[largest]:
        largest = left

    # Checking if right child exists and is greater than largest
    
    if right < n and arr[right] > arr[largest]:
        largest = right

    # If largest is not root, swap and continue heapifying
    
    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]
        heapify(arr, n, largest)


def heapsort(arr):
    
    #Sorting an array using the Heapsort algorithm.
    
    n = len(arr)

    # Step 1: Build max heap
    
    for i in range(n // 2 - 1, -1, -1):
        heapify(arr, n, i)

    # Step 2: Extracting elements from heap one by one
    
    for i in range(n - 1, 0, -1):
        # Move current maximum element to the end
        
        arr[i], arr[0] = arr[0], arr[i]

        # Restore heap property for remaining heap
        heapify(arr, i, 0)

    return arr


# Example usage
data = [12, 11, 13, 5, 6, 7]

print("Original Array:", data)

heapsort(data)

print("Sorted Array:", data)
```

# Empirical Comparison

To evaluate the practical performance of Heapsort, it was compared with Quicksort and Merge Sort using different input sizes and data distributions.

The experiments tested:

* Randomly generated arrays
* Already sorted arrays
* Reverse sorted arrays

The execution time was measured using Python’s time module.

```python
import random
import time

# Quicksort implementation

def quicksort(arr):
    
    # Base case - single element is already sorted
    if len(arr) <= 1:
        return arr
    
    # Taking middle element as pivot. 
    # Much better than taking the first element to avoid O(n^2) worst case on sorted arrays.
    pivot = arr[len(arr)//2]
    
    # Splitting the array into three parts. 
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    
    # Recursively sort and combine.
    return quicksort(left) + middle + quicksort(right)

# Merge Sort implementation
# The classic Divide and Conquer. Standard O(N log N) algorithm. 

def mergesort(arr):
    if len(arr) <= 1:
        return arr
        
    mid = len(arr)//2
    
    # Divide step
    left = mergesort(arr[:mid])
    right = mergesort(arr[mid:])
    
    # Conquer/Merge step
    result = []
    i = j = 0
    
    # Two pointer approach to merge sorted halves
    
    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
            
    
    result.extend(left[i:])
    result.extend(right[j:])
    
    return result

# Utility function
def measure_time(sort_function, data):
    start = time.time()
    
 
    sort_function(data.copy())
    
    end = time.time()
    return end - start

# Testing with different N values to show full analysis
sizes = [1000, 5000, 10000, 50000] 

for size in sizes:
    # Generating datasets for Best, Average, and Worst case scenarios
    random_data = random.sample(range(size*10), size)
    sorted_data = list(range(size))
    reverse_data = list(range(size,0,-1))
    
    print("\nInput Size:", size)

    for name, algorithm in [
        ("Heapsort", heapsort), 
        ("Quicksort", quicksort),
        ("Merge Sort", mergesort)
    ]:
        print(name)
        
        print("Random:",
              measure_time(algorithm, random_data))
        print("Sorted:",
              measure_time(algorithm, sorted_data))
        print("Reverse:",
              measure_time(algorithm, reverse_data))
```

Observed Results and Discussion

After running the code for all three algorithms (Heapsort, Quicksort, and Merge Sort), we can clearly observe that as the value of N (input size) increases, the execution time also increases. This completely matches our theoretical  O(n log n) time complexity. However, when we look closely at the output, there are major differences in how these three perform in real life across random, sorted, and reverse-sorted arrays.
1. Heapsort

    Heapsort is highly consistent. No matter what input we throw at it random, already sorted, or completely reversed the execution time remains almost the same. For example, at the maximum input size of 50,000 elements, it took around 0.176 seconds for reverse data and 0.198 seconds for sorted data. The variation is very minimal, which proves that Heapsort does not care about the initial arrangement of data. But the main issue is that Heapsort was the slowest algorithm in our entire experiment. This is because heap operations (like heapify) require too many comparisons and pointer movements, and it has very poor cache locality compared to the other two.
    
2. Quicksort

    Quicksort absolutely dominated the test and came out as the fastest algorithm for every single input size and distribution. For N=50,000, it finished in just 0.077 seconds for random data, 0.056 seconds for sorted data, and 0.055 seconds for reverse data. Usually, Quicksort gets stuck in a worst-case scenario for sorted inputs, but because we used a smart pivot selection strategy (choosing the middle element), we completely avoided the dreaded O(n^2) pitfall. Because of its highly efficient partitioning and excellent cache utilization, it gave the best practical performance.
    
3. Merge Sort

    Merge Sort safely sits right in the middle—faster than Heapsort but slower than Quicksort. For 50,000 elements, its time ranged from 0.068 seconds (sorted) to 0.113 seconds (random). It gives highly predictable timing, which perfectly aligns with its theoretical O(n log n) nature. The only major disadvantage here is that it requires extra auxiliary space to allocate temporary arrays during the merge step, leading to memory overhead.

Relationship to Theoretical Analysis:

	Heapsort maintained a flat, stable graph regardless of the data ordering, supporting its guaranteed  O(n log n)  performance in best, average, and worst cases.
	Merge Sort also showed highly stable timings across the board, proving its strict mathematical complexity.
	Quicksort gave the lowest execution time everywhere. Even though the theoretical worst-case of Quicksort is O(n^2), we did not see it here because our code creates well-balanced partitions.
    
Conclusion:
Hence, from the above empirical evaluation, it is proved that Quicksort is the best choice for practical speed, Merge Sort gives a good balance of speed and predictability (but wastes RAM), and Heapsort is the safest option for consistent worst-case scenarios even if it sacrifices some speed due to heap overhead.
